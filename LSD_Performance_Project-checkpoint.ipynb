```python
#Importing of Packages & Environemnt setting
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Improting of dataset 
data = pd.read_csv('lsd_math_score_data.csv')
print (data)
```

```python
#Creating dataframe for each column in the dataset
time = data[['Time_Delay_in_Minutes']]
LSD = data[['LSD_ppm']]
score = data[['Avg_Math_Test_Score']]

time_plot = data['Time_Delay_in_Minutes']
LSD_plot = data['LSD_ppm']
```

```python
type(score)
```

```python
# Plotting a line graph 

%matplotlib inline

plt.title('Tissue concentration of LSD over time', fontsize=15)
plt.xlabel('Time (min)', fontsize=14)
plt.ylabel('LSD (ppm)', fontsize=14)
plt.style.use('fivethirtyeight')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.text(x=0, y=-0.7, s='Wagner et al.(1968)', fontsize=14)

plt.plot(time_plot, LSD_plot, color='r', linewidth=3)
plt.show()
```

```python
#Linear Regression

regr = LinearRegression()
regr.fit(LSD, score)
print('Thetal:', regr.coef_[0][0])
print('Intercept:', regr.intercept_[0])
print('R-square:', regr.score(LSD, score))
predicted_score=regr.predict(LSD)
```

```python
#Linear Regression Model Plot 

%matplotlib inline

plt.style.use('fivethirtyeight')
plt.title('Arithmetic Vs LSD-25', fontsize=17)
plt.xlabel('Tissue LSD(ppm)', fontsize=14)
plt.ylabel('Performance Score', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(1, 6.5)
plt.ylim(25, 85)

plt.scatter(LSD, score, s=100, alpha=0.7)
plt.plot(LSD_plot, predicted_score, color='red', linewidth=3)
plt.show()
```
